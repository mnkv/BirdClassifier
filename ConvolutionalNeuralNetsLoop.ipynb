```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle 
import numpy as np
from tensorflow.keras.callbacks import TensorBoard
import time
import os

print(tf.__version__)
#heck that tensorflow is running on GPU
if tf.test.gpu_device_name():
    print('Default GPU Device: {}'.format(tf.test.gpu_device_name()))
else:
    print("Please install GPU version of TF")




```

```python
#limit GPU usage 
gpus = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
```

```python
#open data

X = pickle.load(open('X.pickle', 'rb'))
y = pickle.load(open('y.pickle', 'rb'))

X = np.array(X) #need to convert to numpy for model.fit
y = np.array(y) #need to convert to numpy for model.fit
```

```python
dense_layers = [0, 1, 2]
layer_sizes = [32, 64, 128]
conv_layers = [1, 2, 3]

#normalising data 
X = X/255.0 #or use tf.image.per_image_standardization() 
```

```python
#running models

for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            NAME = '{}-conv-{}-nodes-{}-dense-{}'.format(conv_layer, layer_size, dense_layer, int(time.time()))
            logdir = r'C:\Users\ksesh\files\Documents\BirdClassifierUK\training\\' + r'\logs\\' + NAME + r'\\'
            tensorboard = tf.keras.callbacks.TensorBoard(log_dir=logdir, profile_batch = 10000) #some windows error occurs this solves it
            
            
            model = Sequential()
            model.add(Conv2D(layer_size, (3, 3), input_shape = X.shape[1:]))
            model.add(Activation('relu'))
            model.add(MaxPooling2D(pool_size = (2,2)))

            
            for l in range(conv_layer - 1):
                model.add(Conv2D(layer_size, (3, 3), input_shape = X.shape[1:]))
                model.add(Activation('relu'))
                model.add(MaxPooling2D(pool_size = (2,2)))
            
            model.add(Flatten())
            for l in range(dense_layer):
                model.add(Dense(layer_size))
                model.add(Activation('relu'))

            model.add(Dense(1))
            model.add(Activation('sigmoid'))

            model.compile(loss = 'binary_crossentropy',
                         optimizer = 'adam', 
                         metrics = ['accuracy']) #for more than 2, use categorical

            model.fit(X, y, batch_size = 32, epochs = 15, validation_split = 0.5, callbacks = [tensorboard])
```
